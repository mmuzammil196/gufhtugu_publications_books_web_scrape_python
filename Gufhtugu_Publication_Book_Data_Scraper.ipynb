```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
headers = {'User-Agent' : 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/103.0.0.0 Safari/537.36'}
```

```python
book_link = []

for x in range(1,2,1):
    r = requests.get(f'https://gufhtugu.com/product-category/upcoming-books/page/{x}', headers = headers, verify = False)
    soup = BeautifulSoup(r.text, 'html.parser')
    book_list = soup.find_all('div', class_ = 'title-wrapper')

    for item in book_list:
        link = item.find('a')
        book_link.append(link.get("href"))

print(len(book_link))
print(book_link)
```

```python
req = requests.get('https://gufhtugu.com/artificial-intelligence/', headers = headers, verify = False)
soup = BeautifulSoup(req.text, 'html.parser')
# print(soup)
# try:
name = soup.find('h1', class_ = 'product-title product_title entry-title').text.strip()
print(name)
desc = soup.find('div', class_ = 'product-short-description').text.strip()
print(desc)
rate = soup.find('div', class_ = 'woocommerce-product-rating').text.strip()
print(rate)
cate = soup.find('span', class_ = 'posted_in').text.strip()
print(cate)
tag = soup.find('span', class_ = 'tagged_as').text.strip()
print(tag)
price = soup.find('span', class_ = 'woocommerce-Price-amount amount').text.strip()
print(price)
# except:
  # name = 'no-name'
```

```python
books_data = [] 
for links in book_link:
    req = requests.get(links, headers = headers, verify = False)
    soup = BeautifulSoup(req.text, 'html.parser')
    try:
        b_name = soup.find('h1', class_ = 'product-title product_title entry-title').text.strip()
    except:
        b_name = 'no-name'
    try:
        description = soup.find('div', class_ = 'product-short-description').text.strip()
    except:
        description = 'no-description'
    try:
        category = soup.find('span', class_ = 'posted_in').text.strip()
    except:
        category = 'no-category'
    try:
        tag = soup.find('span', class_ = 'tagged_as').text.strip()
    except:
        tag = 'no-tag'
    try:
        rating = soup.find('div', class_ = 'woocommerce-product-rating').text.strip()
    except:
        rating = 'no-rating'

    books = {
        'Book_name'  : b_name,
        'Description': description,
        'Category' : category,
        'Tags'   : tag,
        'Rating' : rating,
    }
    
    books_data.append(books)
```

```python
df = pd.DataFrame(books_data)
df
```

```python
df.to_csv('upcoming__books.csv', index = False)
```
